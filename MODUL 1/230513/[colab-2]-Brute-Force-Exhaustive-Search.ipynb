```python
def selection_sort(array):
    n = len(array)
   
    for step in range(n):
        min_idx = step

        for i in range(step + 1, n):
         
            if array[i] < array[min_idx]:
                min_idx = i
            
        array[step], array[min_idx] = array[min_idx], array[step]
    
    return array

arr = [7, 2, 5, 4, 17]
selection_sort(arr)
```

```python
def bubble_sort(array):
    n = len(array)
   
    for step in range(n):

        for i in range(0, n-step-1):
         
            if array[i] > array[i+1]:
                temp = array[i]
                array[i] = array[i+1]
                array[i+1] = temp
    
    return array

arr = [7, 2, 5, 4, 17]
bubble_sort(arr)
```

```python
def permute(nums):
    L = []
    n = len(nums)
    i = 0
    while i < n:
        if not L:
            L.append([nums[i]])
        else:
            newl = []
            for elem in L:
                for j in range(0, i+1): 
                    e = elem[:]
                    e.insert(j, nums[i])
                    newl.append(e)
            L = newl
        i += 1

    return L
S = [1, 2, 3]
permute(S)
```

```python
from sys import maxsize
from itertools import permutations

def travelling_salesman_problem(distances, s, num_cities):
    cities = []
    for i in range(num_cities):
        if i != s:
            cities.append(i)
    min_path = maxsize
    next_permutation = permutations(cities)
    for i in next_permutation:
        current_pathweight = 0
        k = s
        for j in i:
            current_pathweight += distances[k][j]
            k = j
        current_pathweight += distances[k][s]
        min_path = min(min_path, current_pathweight)

    return min_path

distances = [
    [0, 2, 8, 5], [2, 0, 3, 4],
    [8, 3, 0, 7], [5, 4, 7, 0]
]
s = 0
num_cities = 4

travelling_salesman_problem(distances, s, num_cities)
```
