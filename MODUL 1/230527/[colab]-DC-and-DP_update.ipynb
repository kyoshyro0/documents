##Divide-and-Conquer

```python
# aivietnam

def merge(S1, S2, S):
    i = j = 0
    while i + j < len(S):
        if j == len(S2) or (i < len(S1) and S1[i] < S2[j]):
            S[i+j] = S1[i]
            i += 1
        else:
            S[i+j] = S2[j]
            j += 1

S = [2, 5, 7, 30, 4, 8]
S1 = [2, 5, 7, 30]
S2 = [4, 8]
merge(S1, S2, S)
S
```

```python
# aivietnam

def merge_sort(S):
    n = len(S)
    
    if n < 2:
        return
    
    mid = n//2
    S1 = S[0:mid]
    S2 = S[mid:n]

    merge_sort(S1)
    merge_sort(S2)

    merge(S1, S2, S)

S = [7, 2, 5, 30, 16, 8, 4, 17]
merge_sort(S)
S
```

```python
# aivietnam

def partition(low, high, S):
    pivot, pointer = S[high], low

    for i in range(low, high):
        if S[i] <= pivot:
            S[i], S[pointer] = S[pointer], S[i]
            pointer += 1

    S[pointer], S[high] = S[high], S[pointer]

    return pointer
```

```python
# aivietnam

def quicksort(low, high, S):
    if len(S) == 1:
        return S

    if low < high:
        p = partition(low, high, S)

        quicksort(low, p-1, S)
        quicksort(p+1, high, S)
 
 
S = [7, 2, 5, 30, 16, 8, 4, 17]
low = 0
high = len(S)-1
quicksort(low, high, S)
S
```

##Dynamic Programming

```python
# aivietnam

def fibonacci_dp(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    
    # create empty dp list (table)
    dp = [0] * (n+1)
    dp[0] = 0
    dp[1] = 1
    for i in range(2, n+1):
        dp[i] = dp[i-1] + dp[i-2]
    
    return dp[n], dp

fibonacci_dp(4)
```

```python
# aivietnam

def longest_subsequence(nums):
    n = len(nums)
    dp = [1] * n
    
    for i in range(n):
        for j in range(i):
            if nums[i] > nums[j]:
                dp[i] = max(dp[i], dp[j] + 1)
    
    return max(dp), dp

nums = [ 3, 10, 2, 1, 20 ]
longest_subsequence(nums)
```

```python
# aivietnam

def longest_subsequence(nums):
    n = len(nums)
    dp = [1] * n
    trace_dp = [-1] * n
    
    for i in range(n):
        for j in range(i):
            if nums[i] > nums[j]:
                # dp[i] = max(dp[i], dp[j] + 1)
                if dp[i] < dp[j] + 1:
                    dp[i] = dp[j] + 1
                    trace_dp[i] = j
    
    result = []
    idx_max = dp.index(max(dp))
    while trace_dp[idx_max] >= 0:
        result.append(nums[idx_max])
        idx_max = trace_dp[idx_max]
    result.append(nums[idx_max])

    return max(dp), dp, sorted(result)

nums = [ 3, 10, 2, 1, 20 ]
longest_subsequence(nums)
```
